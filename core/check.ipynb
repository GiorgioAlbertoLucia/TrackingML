# Test set

```python
import numpy as np
import pandas as pd

test_check = np.load('../data/save/test_check_set.npy')
test_check_df = pd.DataFrame(test_check, columns=['true', 'pred', 'out'])

pos_subset = test_check_df[test_check_df['true'] == 1]
neg_subset = test_check_df[test_check_df['true'] == 0]
test_check_df.describe()
```

```python
len(pos_subset)/ len(test_check_df)
```

```python
# plot the distribution of the true and predicted values
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.hist(pos_subset['out'], bins=100, alpha=0.5, label='pos')
plt.hist(neg_subset['out'], bins=100, alpha=0.5, label='neg')

ax.set_yscale('log')
plt.legend()
```

# Train set

```python
train_check = np.load('../data/save/train_check_set.npy')
train_check_df = pd.DataFrame(train_check, columns=['true', 'pred', 'out'])

pos_train_subset = train_check_df[train_check_df['true'] == 1]
neg_train_subset = train_check_df[train_check_df['true'] == 0]
train_check_df.describe()
```

```python
# plot the distribution of the true and predicted values
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.hist(pos_train_subset['out'], bins=100, alpha=0.5, label='pos')
plt.hist(neg_train_subset['out'], bins=100, alpha=0.5, label='neg')

ax.set_yscale('log')
plt.legend()
```
